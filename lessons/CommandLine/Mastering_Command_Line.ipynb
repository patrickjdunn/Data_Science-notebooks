Command Line Prompt
The command line prompt is a text-based interface used to interact with the operating system. It allows users to execute commands, run programs, and manage files and directories. The command line prompt is powerful and efficient, especially for repetitive tasks, automation, and scripting.
Basics of the Command Line Prompt
Opening the Command Line
•	Windows: You can open the Command Prompt by typing cmd in the search bar or Run dialog (Win + R).
•	Mac: Open the Terminal application found in Applications > Utilities.
•	Linux: Open the Terminal from the applications menu or by pressing Ctrl + Alt + T.
Structure of the Command Line
A typical command line prompt might look something like this:
username@hostname:~$
•	username: Your user name.
•	hostname: The name of the computer.
•	~: The current directory (home directory in this case).
•	$: The prompt symbol (varies by system and user privileges, # for root user).

File Operations
•	touch: Create a New File
touch filename
Creates an empty file with the specified name.
•	cp: Copy Files or Directories
cp source_file destination
Copies a file to the specified destination.
cp -r source_directory destination
Copies a directory and its contents to the specified destination.
•	mv: Move or Rename Files or Directories
mv old_name new_name
Renames a file or directory.
mv file_name /path/to/destination
Moves a file to the specified destination.
•	rm: Remove Files or Directories
rm filename
Deletes the specified file.
rm -r directory
Deletes the specified directory and its contents.


Viewing and Editing Files
•	cat: Concatenate and Display File Content
cat filename
Displays the content of the file.
•	nano: Text Editor
nano filename
Opens the file in the nano text editor.
•	vi: Text Editor
vi filename
Opens the file in the vi text editor.


System Information and Management
•	uname: Display System Information
uname -a
Shows detailed information about the system.
•	top: Display Running Processes
top
Shows a dynamic view of system processes.
•	df: Display Disk Space Usage
df -h
Shows disk space usage in a human-readable format.
•	free: Display Memory Usage
free -h
Shows memory usage in a human-readable format.
