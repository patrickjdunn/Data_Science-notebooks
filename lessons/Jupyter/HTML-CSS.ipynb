A Minimal HTML Document

## A1. A basic HTML page

Every HTML page has:
- `<!DOCTYPE html>`
- `<html>`
- `<head>`
- `<body>`

```python

html = """
<!DOCTYPE html>
<html lang="en">
<head>
  <meta charset="UTF-8">
  <title>My First Web Page</title>
</head>
<body>
  <h1>Hello, Web!</h1>
  <p>This page was created in GitHub Codespaces.</p>
</body>
</html>
"""

with open("index.html", "w") as f:
    f.write(html)

!ls -lh index.html
```

## A2. Common HTML elements

- Headings: `<h1>` â€¦ `<h6>`
- Paragraphs: `<p>`
- Links: `<a>`
- Images: `<img>`
- Lists: `<ul>`, `<ol>`, `<li>`


```python
html = """
<!DOCTYPE html>
<html>
<head>
  <title>HTML Elements</title>
</head>
<body>

<h1>Health Dashboard</h1>

<p>This page summarizes daily health metrics.</p>

<h2>Links</h2>
<a href="https://www.heart.org" target="_blank">American Heart Association</a>

<h2>List</h2>
<ul>
  <li>Blood Pressure</li>
  <li>Heart Rate</li>
  <li>Physical Activity</li>
</ul>

<h2>Image</h2>
<img src="https://upload.wikimedia.org/wikipedia/commons/4/4f/Iconic_image_of_heart.png"
     alt="Heart icon"
     width="150">

</body>
</html>
"""

with open("elements.html", "w") as f:
    f.write(html)
```

```python
html = """
<!DOCTYPE html>
<html>
<body>

<h2>Blood Pressure Readings</h2>

<table border="1">
  <tr>
    <th>Date</th>
    <th>SBP</th>
    <th>DBP</th>
  </tr>
  <tr>
    <td>2026-02-01</td>
    <td>128</td>
    <td>82</td>
  </tr>
  <tr>
    <td>2026-02-02</td>
    <td>142</td>
    <td>90</td>
  </tr>
</table>

</body>
</html>
"""

with open("table.html", "w") as f:
    f.write(html)
```

Inline vs Internal vs External CSS

Weâ€™ll use external CSS (best practice).

```python
css = """
body {
  font-family: Arial, sans-serif;
  background-color: #f5f7fa;
  margin: 40px;
}

h1 {
  color: #b30000;
}

p {
  font-size: 16px;
}

table {
  border-collapse: collapse;
}

th, td {
  padding: 8px 12px;
}
"""

with open("styles.css", "w") as f:
    f.write(css)
```

```python
html = """
<!DOCTYPE html>
<html>
<head>
  <title>Styled Page</title>
  <link rel="stylesheet" href="styles.css">
</head>
<body>

<h1>Daily Health Summary</h1>
<p>This page demonstrates HTML + CSS.</p>

<table border="1">
  <tr>
    <th>Metric</th>
    <th>Value</th>
  </tr>
  <tr>
    <td>Steps</td>
    <td>8,432</td>
  </tr>
  <tr>
    <td>Sleep (hrs)</td>
    <td>7.2</td>
  </tr>
</table>

</body>
</html>
"""

with open("styled.html", "w") as f:
    f.write(html)
```

CSS Selectors & Classes
ðŸ§© Classes (very important)

```python
.card {
  background: white;
  padding: 20px;
  border-radius: 8px;
}
```

```python
html = """
<!DOCTYPE html>
<html>
<head>
<link rel="stylesheet" href="styles.css">
<style>
.card {
  background: white;
  padding: 20px;
  margin-bottom: 20px;
  border-radius: 8px;
}
</style>
</head>
<body>

<h1>Patient Overview</h1>

<div class="card">
  <h2>Vitals</h2>
  <p>BP: 128/82</p>
  <p>HR: 72</p>
</div>

<div class="card">
  <h2>Activity</h2>
  <p>Steps: 8,432</p>
  <p>Minutes: 42</p>
</div>

</body>
</html>
"""

with open("cards.html", "w") as f:
    f.write(html)
```

```python
html = """
<!DOCTYPE html>
<html>
<head>
  <title>My Health Dashboard</title>
  <style>
    body { font-family: Arial; background: #eef2f7; margin: 40px; }
    .section { background: white; padding: 20px; margin-bottom: 20px; border-radius: 8px; }
  </style>
</head>
<body>

<h1>My Health Dashboard</h1>

<div class="section">
  <h2>Vitals</h2>
  <p>BP:</p>
  <p>HR:</p>
</div>

<div class="section">
  <h2>Activity</h2>
  <p>Steps:</p>
  <p>Sleep:</p>
</div>

</body>
</html>
"""

with open("dashboard.html", "w") as f:
    f.write(html)
```
